````
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt

# Datos de ejemplo
data = pd.DataFrame({'Sex': titanic.loc[:, "Sex"], 'Survived': titanic.loc[:, "Survived"]}).replace({1: "Survived", 0: "Died"})
tita = data.sort_values('Sex')
color_palette = {'Survived': 'green', 'Died': 'red'}  # Paleta de colores personalizada

mosaic(tita, ['Sex', 'Survived'], gap=0.007, labelizer=lambda k: "")  # Elimina las etiquetas para los subgráficos
plt.title('Sex vs Survived', fontsize=18)

# Agregar leyenda personalizada
legend_labels = ['Survived', 'Died']
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_palette[label], markersize=10) for label in legend_labels]
plt.legend(handles, legend_labels, title='Survival Status', loc='upper right')

plt.show()
````

# Interpretation

This Mosaic reveals a  gender-based contrast: women exhibited a notably higher survival rate compared to men. This trend is likely from the emergency protocol prioritizing women's evacuation, leading to a significantly higher mortality rate among men—more than double that of women. This outcome underscores the impact of evacuation priorities, resulting in a distinct survival pattern between genders during the Titanic's crisis.

````
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic

# Datos de ejemplo
tita1 = titanic.sort_values('Pclass')

# Paleta de colores personalizada
color_palette = {'Class 3': 'blue', 'Class 2': 'green','Class 1': 'red'}

mosaic(tita1, ['Pclass', 'Survived'], gap=0.01, labelizer=lambda k: "")
plt.title('Survived vs Pclass', fontsize=18)

# Agregar leyenda personalizada
legend_labels = ['Class 3', 'Class 2','Class 1']
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_palette[label], markersize=10) for label in legend_labels]
plt.legend(handles, legend_labels, title='Survival Status', loc='upper right')

plt.show()
````

# Interpretation

This mosaic shows the role of social class in the passengers survival dynamics. It becomes evident that along with gender, stratification played an important role in evacuation decisions. First-class passengers, associated with higher status, exhibited a remarkably higher survival rate. On the contrary
, the third-class passengers, representing lower purchasing power, faced the highest mortality rate. This underscores the impact of socioeconomic position on access to resources during the Titanic's evacuation, emphasizing the complexities of class-based disparities in survival outcomes.


````
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic

# Datos de ejemplo
tita2 = titanic.sort_values('Embarked')

# Paleta de colores personalizada
color_palette = {'S': 'blue', 'Q': 'green','C': 'red'}

mosaic(tita2, ['Embarked', 'Survived'], gap=0.005, labelizer=lambda k: "")
plt.title('Survived vs Embarked', fontsize=18)

# Agregar leyenda personalizada
legend_labels = ['S', 'Q','C']
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_palette[label], markersize=10) for label in legend_labels]
plt.legend(handles, legend_labels, title='Survival Status', loc='upper right')

plt.show()
````

# Interpretation

This mosaic shows a strong connection between embarkation points and survival rates. Passengers from Cherbourg experienced the highest survival rate, possibly due to a concentration of women or first-class individuals. In contrast, Southampton, with the highest mortality rate, likely saw more men or lower-class passengers.
